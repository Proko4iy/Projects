```python
import pandas as pd
import missingno as msno 
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

import datetime as dt
```

```python
df=pd.read_excel('Когортный анализ.xlsx')
```

```python
df.head()
```

```python
msno.bar(df)
```

```python
missing_values = ((df.isna().sum() / len(df)) * 100).sort_values()
```

```python
missing_values
```

```python
def get_month(x):

    return dt.datetime(x.year, x.month, 1)

df['month_year'] = df.date.apply(get_month)
```

```python
df.head()
```

```python
df['cohortMonth']=df.groupby('user_id')['month_year'].transform('min')
```

```python
df.head(50)
```

```python
def get_date(df,column):
    year=df[column].dt.year
    month=df[column].dt.month
    day=df[column].dt.day
    return year,month,day
```

```python
date_year, date_month,_ =get_date(df,'date')
```

```python
cohort_year,cohort_month,_=get_date(df,'cohortMonth')
```

```python
year_diff=date_year-cohort_year
month_diff=date_month-cohort_month
```

```python
df['cohort_index']=year_diff*12+month_diff+1
```

```python
df.head()
```

```python
df['cohort_index'].unique()
```

```python
df1=df.pivot_table(index='cohortMonth',columns='cohort_index',values='user_id',aggfunc=len)
```

```python
df1
```

```python
df2=df.pivot_table(index='cohortMonth',columns='cohort_index',values='price',aggfunc=sum)
```

```python
df2
```

```python
first_column=df1.iloc[:,0]
```

```python
retention1=df1.div(first_column,axis=0)
retention1.round(2)
```

```python
first_column=df2.iloc[:,0]
retention2=df2.div(first_column,axis=0)
retention2.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ - retention')
sns.heatmap(data=retention1,annot=True,fmt='.0%',vmin=0.0,vmax=0.5,cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ - retention')
sns.heatmap(data=retention2,annot=True,fmt='.0%',vmin=0.0,vmax=0.5,cmap="YlGnBu")
plt.show()
```
