```python
test_loss=test[~test['loss_payout_amt'].isna()]
```

```python
control_loss=control[~control['loss_payout_amt'].isna()]
```

## Видим что средняя цена полиса в тестовой группе выросла что явлется хорошим сигналом

## Рассмотрим остальные метрики с помощью проверки гипотез

```python
t_statistic, p_value = ttest_ind(control['conversion to checkout'],test['conversion to checkout'] , equal_var=False)
alpha = 0.05
if p_value < alpha:
    if t_statistic > 0:
        print("Среднее значение первой выборки статистически значимо больше среднего значения второй выборки")
    else:
        print("Среднее значение второй выборки статистически значимо больше среднего значения первой выборки")
else:
    print("Нет статистически значимых различий между средними значениями выборок")
```

```python
t_statistic, p_value = ttest_ind(control_loss['conversion to loss'],test_loss['conversion to loss'] , equal_var=False)
alpha = 0.05
if p_value < alpha:
    if t_statistic > 0:
        print("Среднее значение первой выборки статистически значимо больше среднего значения второй выборки")
    else:
        print("Среднее значение второй выборки статистически значимо больше среднего значения первой выборки")
else:
    print("Нет статистически значимых различий между средними значениями выборок")
```
